# K-Nearest Neighbours

```python
import pandas as pd
df = pd.read_csv("data/fashion_mnist.csv")
df.head() 
```

```python
label_map = {
    0: "T-shirt/top",
    1: "Trouser",
    2: "Pullover",
    3: "Dress",
    4: "Coat",
    5: "Sandal",
    6: "Shirt",
    7: "Sneaker",
    8: "Bag",
    9: "Ankle boot",
}
```

```python
df.info()
```

```python
import numpy as np
import matplotlib.pyplot as plt

my_random_seed = 100

numbers_to_display = 36
num_cells = int(np.ceil(np.sqrt(numbers_to_display)))
np.random.seed(my_random_seed)
sample_index_list = np.random.choice(list(df.index), numbers_to_display)

plt.figure(figsize=(10, 10))
for plot_index in range(numbers_to_display):    
    sample_index = sample_index_list[plot_index]
    plt.subplot(num_cells, num_cells, plot_index + 1)
    frame = df.loc[sample_index,df.columns!="label"].to_numpy().reshape(28, 28)
    plt.imshow(frame, cmap="Greys")
    gt_label = label_map.get(df.iloc[sample_index,0])
    plt.title(gt_label)
    plt.tick_params(
        axis="both",
        which="both",
        bottom=False,
        left=False,
        labelbottom=False,
        labelleft=False,
    )

plt.subplots_adjust(hspace=0.5, wspace=0.5)
plt.show()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    df.iloc[:, 1:].to_numpy(),
    df.iloc[:, 0].to_numpy().flatten(),
    test_size=0.4,
    random_state=my_random_seed,
    stratify=df.iloc[:,0].to_numpy(),
)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(
    n_neighbors=5, p=2, metric="minkowski", weights="uniform", algorithm="auto"
)
knn.fit(x_train, y_train)
y_pred_test_knn = knn.predict(x_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_test_knn))
```
