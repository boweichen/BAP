## Data preparation

i) encode the categorical features into a set of binary/dummy variables; ii) encode the response variable to be the dummy variable (0 - Bad, 1 - Good)

**Hint**: 

* Method 1: We can use `pandas.get_dummies` to encode category features and response variable 
https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.get_dummies.html

* Method 2: we can use OneHotEncoder from sklearn to encode categorical features.

```python
df2 = pd.get_dummies(df, drop_first=True)
df2.head(5)
```

```python
x = df2.iloc[:, :-1].to_numpy()
y = df2.iloc[:, -1].to_numpy()
print(x)
print(y)
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler().fit(x[:, :2])
x[:, :2] = ss.transform(x[:, :2])
```

Please split the data into training and test sets, test size is 0.4.

```python
from sklearn.model_selection import train_test_split
my_random_seed_number = 100
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.4, random_state=my_random_seed_number, stratify=y
)
```

## MLP

```python
from sklearn.neural_network import MLPClassifier

clf = MLPClassifier(random_state=my_random_seed_number, max_iter=5000)
clf.fit(x_train, y_train)
```

## Model evaluation 

Make predictions with both the training data and the test data.

```python
y_pred_label_train = clf.predict(x_train)
y_pred_label_prob_train = clf.predict_proba(x_train)[:, 1]

y_pred_label_test = clf.predict(x_test)
y_pred_label_prob_test = clf.predict_proba(x_test)[:, 1]
```

Create two data frames. Each data frame has three columns:
    - Label_GT (ground truth in the dataset)
    - Label_Pred_Prob (probability of your model prediction)
    - Label_Pred (your model predicted label)

```python
df_result_train = pd.DataFrame([y_train, y_pred_label_prob_train, y_pred_label_train])
df_result_train = df_result_train.T
df_result_train.columns = ["Label_GT", "Label_Pred_Prob", "Label_Pred"]
df_result_train.head()
```

```python
df_result_test = pd.DataFrame([y_test, y_pred_label_prob_test, y_pred_label_test])
df_result_test = df_result_test.T
df_result_test.columns = ["Label_GT", "Label_Pred_Prob", "Label_Pred"]
df_result_test.head()
```

Use `confusion_matrix` to calculate the TP, FP, TN, FN of both the training data and the test data.

```python
from sklearn.metrics import confusion_matrix

print("Training")
print(confusion_matrix(y_train, y_pred_label_train))
print("--\n")

print("Test")
print(confusion_matrix(y_test, y_pred_label_test))
```

Calculate the related evaluation metrics (accuracy, recall, precision) for both the training and the test data.

```python
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score

print("Compute evaluation metrics (training)")
print("Average accuracy: %0.4f" % clf.score(x_train, y_train))
print(
    "Average recall: %0.4f" % recall_score(y_train, y_pred_label_train, average="macro")
)
print(
    "Average precision: %0.4f"
    % precision_score(y_train, y_pred_label_train, average="macro")
)

print("Compute evaluation metrics (test)")
print("Average accuracy: %0.4f" % clf.score(x_test, y_test))
print(
    "Average recall: %0.4f" % recall_score(y_test, y_pred_label_test, average="macro")
)
print(
    "Average precision: %0.4f"
    % precision_score(y_test, y_pred_label_test, average="macro")
)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_train, y_pred_label_train))

print(classification_report(y_test, y_pred_label_test))
```

Use `roc_curve` and `auc` from `sklearn.metrics` to calculate the AUC of your model performance in the test data, and define a function to draw a ROC graph of your model

```python
from sklearn.metrics import roc_curve, auc

fpr, tpr, thresholds = roc_curve(y_test, y_pred_label_prob_test)
roc_auc = auc(fpr, tpr)
```

```python
def plot_roc_curve(fpr, tpr):
    fig, ax = plt.subplots(figsize=(5, 5))
    ax.step(fpr, tpr, color="Orange", label="ROC curve (area = %0.4f)" % roc_auc)
    ax.plot([0, 1], [0, 1], color="RoyalBlue", linestyle="--")
    ax.set_xlim([0.0, 1.0])
    ax.set_ylim([0.0, 1.0])
    ax.set_xlabel("False Positive Rate")
    ax.set_ylabel("True Positive Rate")
    ax.set_title("Receiver Operating Characteristic (ROC) Curve")
    ax.legend(loc="lower right")
    plt.show()
    
plot_roc_curve(fpr, tpr)
```
