# Tree-Based Models | Exercise Solution

```python
import pandas as pd
```

```python
df = pd.read_csv("data/bank_marketing.csv", sep=";")
```

```python
df.head(3)
```

```python
df.info()
```

```python
df["y"].value_counts()
```

```python
df.isnull().sum()
```

```python
from sklearn.utils import resample

my_random_seed = 100

# Separate majority and minority classes
df_majority = df.loc[df["y"] == "no", :]
df_minority = df.loc[df["y"] == "yes", :]
```

```python
# Downsampling
df_majority_downsampled = resample(
    df_majority,
    replace=False,  # sample with replacement
    n_samples=df_minority.shape[0],  # to match majority class
    random_state=my_random_seed,
)
df_majority_downsampled
```

```python
df_downsampled = pd.concat([df_majority_downsampled, df_minority])
df_downsampled
```

```python
df_downsampled.reset_index(drop=True, inplace=True)
df_downsampled
```

```python
df_downsampled["y"].value_counts()
```

```python
df_raw = df.copy()
df = df_downsampled.copy()
```

```python
df.info()
```

```python
df = pd.get_dummies(data=df, drop_first=True, dtype="int8")
df.head()
```

```python
feature_name = df.columns.tolist()
```

```python
feature_name.remove("y_yes")
feature_name
```

```python
df[feature_name].head()
```

```python
x = df[feature_name].to_numpy()
x
```

```python
y = df["y_yes"]
```

```python
from sklearn.model_selection import train_test_split

split_train_size = 0.7 # [0.5, 0.9]

x_train, x_test, y_train, y_test = train_test_split(
    x, y, train_size=split_train_size, random_state=my_random_seed, stratify=y
)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
import numpy as np

dtc_param = {
    "criterion": ["entropy", "gini"],
    "max_depth": np.arange(3, 10).tolist(),
    "min_samples_leaf": np.arange(3, 10).tolist(),
}

n_folds = 5

dtc = DecisionTreeClassifier(random_state=my_random_seed)

dtc_cv = GridSearchCV(
    estimator=dtc, param_grid=dtc_param, cv=n_folds, scoring="accuracy"
)
```

```python
dtc_cv.fit(x_train, y_train)
```

```python
dtc = DecisionTreeClassifier(random_state=my_random_seed, **dtc_cv.best_params_)
dtc.fit(x_train, y_train)
```

```python
label_names = df["y_yes"].value_counts().index.tolist()
label_names = [str(x) for x in label_names]
label_names
```

```python
from sklearn.tree import export_graphviz
import pydotplus
from IPython.display import Image

dot_data = export_graphviz(
    dtc,
    out_file=None,
    feature_names=feature_name,
    class_names=label_names,
    filled=True,
    rounded=True,
    special_characters=True,
)

# Draw graph
graph = pydotplus.graph_from_dot_data(dot_data)

# Show graph
Image(graph.create_png())
```

```python
y_test_pred = dtc.predict(x_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_test_pred))
```

```python
from sklearn.inspection import permutation_importance

fpi_dtc = permutation_importance(dtc, x_train, y_train, random_state=my_random_seed)
```

```python
print("feature_importance: {}".format(fpi_dtc.importances_mean))

feature_importance_index = fpi_dtc.importances_mean.argsort()
print("feature_importance_index: {}".format(feature_importance_index))

print("feature_column: {}".format(feature_name))
```

```python
df_fpi_dtc = pd.DataFrame(
    {"Feature": feature_name, "Importance Score": fpi_dtc.importances_mean}
)

df_fpi_dtc.sort_values("Importance Score", ascending=False)
```

```python
import eli5

eli5.show_weights(dtc, feature_names=feature_name)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_param = {
    "criterion": ["entropy", "gini"],
    "max_depth": np.arange(3, 10).tolist(),
    "min_samples_leaf": np.arange(3, 10).tolist(),
}

rf = RandomForestClassifier(random_state=my_random_seed)

rf_cv = GridSearchCV(estimator=rf, param_grid=rf_param, cv=n_folds, scoring="accuracy")
```

```python
rf_cv.fit(x_train, y_train)
```

```python
rf = RandomForestClassifier(random_state=my_random_seed, **rf_cv.best_params_)
rf.fit(x_train, y_train)
```

```python
y_test_pred_rf = rf.predict(x_test)
print(classification_report(y_test, y_test_pred_rf))
```

```python
import eli5

eli5.show_weights(rf, feature_names=feature_name)
```
