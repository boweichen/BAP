# Tree-Based Methods

```python
import pandas as pd

df = pd.read_excel("data/corporate_rating_moody.xlsx")
df.info()
```

```python
df["Rating"].value_counts()
```

```python
# Function to map ratings to categories
def categorize_rating(rating):
    if rating in ["AAA", "AA"]:
        return "High quality (AAA/AA)"
    elif rating == "A":
        return "Upper-medium (A)"
    elif rating == "BBB":
        return "Medium (BBB)"
    elif rating == "BB":
        return "Speculative (BB)"
    elif rating == "B":
        return "Vulnerable (B)"
    elif rating in ["C", "CC", "CCC"]:
        return "Poor standing (CCC/CC/C)"
```

```python
df["Rating category"] = df["Rating"].apply(categorize_rating)
```

```python
df["Rating category"].value_counts()
```

```python
del df["Rating"]
```

```python
del df["Rating agency name"]
```

```python
df.head()
```

```python
from sklearn.utils import resample

# Determine the most frequent category
most_frequent_category = df["Rating category"].value_counts().idxmax()
```

```python
most_frequent_category
```

```python
# Separate the DataFrame by category
dfs = {
    category: df[df["Rating category"] == category]
    for category in df["Rating category"].unique()
}
```

```python
# Upsample each category to match the most frequent category's count

random_seed_val = 100
upsampled_dfs = [
    resample(
        dfs[category],
        replace=True,
        n_samples=df["Rating category"].value_counts()[most_frequent_category],
        random_state=random_seed_val,
    )
    for category in dfs
    if category != most_frequent_category
]
```

```python
# Append the most frequent category DataFrame as is
upsampled_dfs.append(dfs[most_frequent_category])

# Concatenate all upsampled DataFrames
balanced_df = pd.concat(upsampled_dfs)
```

```python
balanced_df.info()
```

```python
balanced_df["Rating category"].value_counts()
```

```python
# one-hot encoding
balanced_df_features = pd.get_dummies(
    balanced_df.iloc[:, :-1], drop_first=True, dtype="int8"
)
balanced_df_features.head()
```

```python
balanced_df_features.info()
```

```python
from sklearn.preprocessing import StandardScaler

x = balanced_df_features.to_numpy()
x[:, :26] = StandardScaler().fit_transform(x[:, :26])
print(x)
```

```python
x.shape
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit(balanced_df["Rating category"])
print(le.classes_)
y = le.transform(balanced_df["Rating category"])
print(y)
```

```python
from sklearn.model_selection import GridSearchCV
import numpy as np

model_cv_obj = {}
model_cv_results = {}
fold_cv = 10
no_jobs = -1


# Random forest
from sklearn.ensemble import RandomForestClassifier
para_rf = {
    "max_depth": [10, 25, 50],
    "min_samples_leaf": [1, 5, 10],
    "min_samples_split": [2, 5, 10],
    "n_estimators": [100, 200, 500],
}
rf = RandomForestClassifier(random_state=random_seed_val)
model_cv_obj["RF"] = GridSearchCV(
    estimator=rf,
    param_grid=para_rf,
    n_jobs=no_jobs,
    cv=fold_cv,
    scoring="accuracy",
)

# AdaBoost
from sklearn.ensemble import AdaBoostClassifier

para_adaboost = {
    "learning_rate": [0.001, 0.01, 0.05, 0.1, 0.15],
    "n_estimators": [100, 200, 500],
}
adaboost = AdaBoostClassifier(random_state=random_seed_val)
model_cv_obj["AdaBoost"] = GridSearchCV(
    estimator=adaboost,
    param_grid=para_adaboost,
    n_jobs=no_jobs,
    cv=fold_cv,
    scoring="accuracy",
)

# GBDT
from sklearn.ensemble import GradientBoostingClassifier

para_gbdt = {
    "learning_rate": [0.001, 0.01, 0.05, 0.1, 0.15],
    "n_estimators": [100, 200, 500],
}
gbdt = GradientBoostingClassifier(random_state=random_seed_val)
model_cv_obj["GBDT"] = GridSearchCV(
    estimator=gbdt,
    param_grid=para_gbdt,
    n_jobs=no_jobs,
    cv=fold_cv,
    scoring="accuracy",
)

# XGBoost
from xgboost import XGBClassifier

para_xgb = {
    "learning_rate": [0.001, 0.01, 0.05, 0.1, 0.15],
    "max_depth": [10, 25, 50],
    "min_child_weight": [2, 3, 4],
    "gamma": [0.0, 0.1, 0.2, 0.3, 0.4],
    "colsample_bytree": [0.3, 0.4, 0.5, 0.7],
}
xgb = XGBClassifier(random_state=random_seed_val)
model_cv_obj["XGBoost"] = GridSearchCV(
    estimator=xgb,
    param_grid=para_xgb,
    n_jobs=no_jobs,
    cv=fold_cv,
    scoring="accuracy",
)

# Decision tree
from sklearn.tree import DecisionTreeClassifier

para_dtc = {
    "criterion": ["entropy", "gini"],
    "max_depth": np.arange(3, 10).tolist(),
    "min_samples_leaf": np.arange(3, 10).tolist(),
}
dtc = DecisionTreeClassifier(random_state=random_seed_val)
model_cv_obj["DT"] = GridSearchCV(
    estimator=dtc, 
    param_grid=para_dtc, 
    n_jobs=no_jobs, 
    cv=fold_cv, 
    scoring="accuracy",
)
```

```python
from sklearn.metrics import accuracy_score

for algo in model_cv_obj.keys():
    print(algo)
    model_cv_obj[algo].fit(x, y)
    model_cv_results[algo] = model_cv_obj[algo].best_score_
```

```python
model_cv_results
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
import seaborn as sns

sns.set_style("whitegrid")

x_plot = [x for y, x in sorted(zip(model_cv_results.values(), model_cv_results.keys()))]
y_plot = [y for y, x in sorted(zip(model_cv_results.values(), model_cv_results.keys()))]
x_pos = [i for i, _ in enumerate(x_plot)]

plt.barh(x_pos, y_plot, color="royalblue")
plt.ylabel("Model")
plt.xlabel("Accuracy")

plt.yticks(x_pos, x_plot)
plt.show()
```

```python
from sklearn.inspection import permutation_importance

fpi_dtc = permutation_importance(
    model_cv_obj["XGBoost"], x, y, random_state=random_seed_val
)
```
